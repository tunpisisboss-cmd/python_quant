```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.graphics import tsaplots
```

```python
gold = yf.download('GC=F',start = '2020-01-01')
gold = gold.xs('GC=F',axis = 1,level =1)

dxy = yf.download('DX-Y.NYB',start = '2020-01-01')
dxy = dxy.xs('DX-Y.NYB',axis = 1,level =1)





```

```python
gold['return'] = gold['Close'].pct_change()*100
dxy['return'] = dxy['Close'].pct_change()*100
```

```python
tsaplots.acf(gold['return'].dropna())
```

```python
plt.show()
```

```python
df = pd.DataFrame({
    'Gold' : gold['return'],
    'DXY' : dxy['return']
}).dropna() # Add .dropna() to remove rows with NaN values

x = sm.add_constant(df['DXY'])
model = sm.OLS(df['Gold'],x).fit()
print(model.summary())
```

```python
df['predicted_gold'] = model.predict()
print(df[['Gold','predicted_gold']].round(2))
```
