```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

gold = yf.download('GC=F', start='2020-01-01')
gold = gold.xs('GC=F', axis=1, level=1)
```

```python
gold['Return'] = gold['Close'].pct_change() * 100
gold['MA50']   = gold['Close'].rolling(50).mean()
gold['MA200']  = gold['Close'].rolling(200).mean()
```

```python
gold[['Close','MA50','MA200']].plot(title='Gold + MA')
plt.show()
```

```python
# สร้างสัญญาณซื้อขาย
gold['Signal'] = 0
gold.loc[gold['MA50'] > gold['MA200'], 'Signal'] = 1
gold.loc[gold['MA50'] < gold['MA200'], 'Signal'] = -1

print(gold['Signal'].value_counts())
```

```python
gold['Strategy_Return'] = gold['Signal'].shift(1) * gold['Return']

gold['Cumulative'] = (1 + gold['Strategy_Return']/100).cumprod()

gold['Cumulative'].plot(title='Backtest Result')
plt.show()
```

```python
sharpe = gold['Strategy_Return'].mean() / gold['Strategy_Return'].std() * (252**0.5)
print('Sharpe Ratio:', round(sharpe, 2))
```

```python
# คำนวณ RSI
delta = gold['Close'].diff()
gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)

avg_gain = gain.rolling(14).mean()
avg_loss = loss.rolling(14).mean()

rs = avg_gain / avg_loss
gold['RSI'] = 100 - (100 / (1 + rs))

# วาดกราฟ RSI
gold['RSI'].plot(title='RSI')
plt.axhline(70, color='red')
plt.axhline(30, color='green')
plt.show()
```

```python
# สร้างสัญญาณ RSI
gold['RSI_Signal'] = 0
gold.loc[gold['RSI'] < 30, 'RSI_Signal'] = 1   # Buy
gold.loc[gold['RSI'] > 70, 'RSI_Signal'] = -1  # Sell

# คำนวณผลกำไร
gold['RSI_Return'] = gold['RSI_Signal'].shift(1) * gold['Return']

# Sharpe Ratio
sharpe_rsi = gold['RSI_Return'].mean() / gold['RSI_Return'].std() * (252**0.5)
print('RSI Sharpe Ratio:', round(sharpe_rsi, 2))
```

```python
# รวม MA + RSI
gold['Combined_Signal'] = 0

# Buy เมื่อ MA50 > MA200 และ RSI < 50
gold.loc[(gold['MA50'] > gold['MA200']) &
         (gold['RSI'] < 50), 'Combined_Signal'] = 1

# Sell เมื่อ MA50 < MA200 และ RSI > 50
gold.loc[(gold['MA50'] < gold['MA200']) &
         (gold['RSI'] > 50), 'Combined_Signal'] = -1

# คำนวณผลกำไร
gold['Combined_Return'] = gold['Combined_Signal'].shift(1) * gold['Return']

# Sharpe Ratio
sharpe_combined = gold['Combined_Return'].mean() / gold['Combined_Return'].std() * (252**0.5)
print('Combined Sharpe Ratio:', round(sharpe_combined, 2))# รวม MA + RSI
gold['Combined_Signal'] = 0

# Buy เมื่อ MA50 > MA200 และ RSI < 50
gold.loc[(gold['MA50'] > gold['MA200']) &
         (gold['RSI'] < 50), 'Combined_Signal'] = 1

# Sell เมื่อ MA50 < MA200 และ RSI > 50
gold.loc[(gold['MA50'] < gold['MA200']) &
         (gold['RSI'] > 50), 'Combined_Signal'] = -1

# คำนวณผลกำไร
gold['Combined_Return'] = gold['Combined_Signal'].shift(1) * gold['Return']

# Sharpe Ratio
sharpe_combined = gold['Combined_Return'].mean() / gold['Combined_Return'].std() * (252**0.5)
print('Combined Sharpe Ratio:', round(sharpe_combined, 2))
```
