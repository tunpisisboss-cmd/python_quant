```python
tickers = ['BTC-USD','GC=F','CL=F']
data = yf.download(tickers,start = '2020-01-01')['Close'].dropna()
```

```python
ma50 = data.rolling(50).mean()
ma200 = data.rolling(200).mean()
```

```python
buy = ma50 > ma200
sell = ma50 < ma200

pf = vbt.Portfolio.from_signals(data,entries=buy,exits=sell,fees=0.001)
```

```python
pf.stats()
```

```python
pf['CL=F'].plot().show()
```

```python
#optimize
fast_window_lengths = [10,20,30]
slow_window_lengths = [50,100,200]

results = {}

for f in fast_window_lengths:
  for s in slow_window_lengths:
    ma_fast = data.rolling(f).mean()
    ma_slow = data.rolling(s).mean()

    buy = ma_fast > ma_slow
    sell = ma_fast < ma_slow

    pf = vbt.Portfolio.from_signals(data,entries=buy,exits=sell,fees=0.001,freq="1D")

    results[f'{f}vs{s}'] = pf.stats()['Sharpe Ratio']
```
