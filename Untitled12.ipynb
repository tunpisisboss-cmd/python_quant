```python
btc  = yf.download('BTC-USD',start = '2020-01-01')
btc  = btc.xs('BTC-USD',axis=1,level=1)
```

```python
ma20 = btc['Close'].rolling(20).mean()
ma50 = btc['Close'].rolling(50).mean()

buy = ma20 > ma50
sell = ma20 < ma50
```

```python
pf= vbt.Portfolio.from_signals(btc['Close'],
                                entries=buy,
                                exits=sell,
                                init_cash=10000,
                                size = 0.1,
                                size_type='percent',
                                fees=0.001
                                )
```

```python
print(pf.stats().round(2))
```

```python
#การแบ่งข้อมูลออกเป็น train และ test
train =  btc['2020':'2023']
test = btc['2024':]
#train ma
ma20 = train['Close'].rolling(20).mean()
ma50 = train['Close'].rolling(50).mean()

buy_train = ma20 > ma50
sell_train = ma20 < ma50
#test ma
ma20 = test['Close'].rolling(20).mean()
ma50 = test['Close'].rolling(50).mean()

buy_test  = ma20 > ma50
sell_test = ma20 < ma50
# backtest on test onlyy
pf_test = vbt.Portfolio.from_signals(test['Close'],entries=buy_test,exits=sell_test,fees=0.001)
```

```python
print(pf_test.stats().round(2))
```
