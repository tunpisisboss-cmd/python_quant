```python
import vectorbt as vbt
import pandas as pd
import yfinance as yf
```

```python
gold = yf.download('GC=F',start = '2020-01-01')
gold = gold.xs('GC=F',axis = 1,level=1)
```

```python
ma50 = gold['Close'].rolling(50).mean()
ma200 = gold['Close'].rolling(200).mean()

buy = ma50 > ma200
sell = ma50 < ma200
```

```python
pf = vbt.Portfolio.from_signals(gold['Close'],entries=buy,exits=sell)
```

```python
print(pf.stats())
```
